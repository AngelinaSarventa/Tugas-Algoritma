```python
#STRINGS INDEX
hobi = "Hera Suka Membuat Kue"
```

```python
print(hobi[0:9])
```

```python
print(hobi[10:-1])
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#-1 adalah huruf terakhir seperti contoh diatas huruf ke 10 adalah M dan huruf terakhir adalah e
#sehingga yang dihasilkan dari perintah print diatas adalah membuat ku karena yang diminta adalah huruf ke 10 sampai -1 
```

```python
print(hobi[:])
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#untuk print semua kalimat cukup menggunakan [:]
```

```python
print(hobi[2:-1:2])
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#Jika ingin print dengan jarak 2 huruf dan dimulai dengan huruf kedua tapi huruf terakhir yaitu e tidak diikutkan 
# maka gunakan perintah seperti di atas
```

```python
#OPERATOR PERBANDINGAN DI PYTHON
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
x=7>5
print(x)
x=5<7
print(x)
x=4 >=3
print(x)
x=3 <=4
print(x)
x=5 ==5
print(x)
x=6 !=5
print(x)
```

```python
#OPERATOR LOGIKA DI PYTHON
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
anak_perempuan = 20
anak_laki2 = 15
print (anak_perempuan > 13 and anak_laki2 < 30)
```

```python
print (anak_perempuan < 13 or anak_laki2 < 30)
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#walaupun ada yang salah tetapi hasilnya akan tetap true karena salah satunya benar
```

```python
print (anak_perempuan < 13 or anak_laki2 > 30)
```

```python
print (not(anak_perempuan < 13 or anak_laki2 > 30))
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#not akan menhasilkan jawaban yang berlawanan maka jawabannya adalah true
```

```python
#OPERATOR IF DI PYTHON
lingkar_badan = float (input("berapakah lingkar badan anda? ")) 
cm_atau_inci = input("apakah dalam cm atau dalam inci? ")

if cm_atau_inci.upper() == "CM":
    lingkar_badan = lingkar_badan*0.393701
    print("lingkar badan anda adalah " + str(lingkar_badan) + " inci")
elif cm_atau_inci.upper() == "INCI":
    lingkar_badan = lingkar_badan/0.393701
    print(f"lingkar badan anda adalah {lingkar_badan} inci")
```
