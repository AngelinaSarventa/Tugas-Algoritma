```python
#WHILE LOOP

a = 0
while a < 7:
    print(a * "*")
    a += 1
#%%%%%%%%%%%%%%%%%%%%%%
#perintah while akan tidak berfungsi atau tidak menghasilkan apapun saat syarat yang diberikan tidak benar
```

```python
a = 0
while a > 7:
    print(a * "*")
    a += 1
#%%%%%%%%%%%%%%%%%%%%%%
#seperti ini adalah contoh syarat yang diberikan salah
```

```python
#LIST DI PYTHON

bulan = ["januari", "februari", "maret", "april", "mei"]
print(bulan)
```

```python
print(bulan[2])
```

```python
print(bulan[-1])
```

```python
print(bulan[0:4])
#%%%%%%%%%%%%%%%%%%%%%%%%%%
#jika pada strings index yang dihitung adalah jumlah huruf beserta spasi disini yang dihitung aalah jumlah kata
```

```python
bulan[0] = "desember"
print(bulan)
#%%%%%%%%%%%%%%%%%%%%%%%%%%
#ketika kita ingin mengubah list yang ada bisa gunakan seperti ini
```

```python
#METODA LIST

bulan = ["januari", "maret", "april"]
bulan.append("mei")
print(bulan)
#%%%%%%%%%%%%%%%%%%%%%%%%%%%
#ketika kita ingin menambahkan isi pada list yang ada bisa gunakan append
```

```python
bulan.insert(1,"februari")
print(bulan)
#%%%%%%%%%%%%%%%%%%%%%%%%%%%
#ketika ingin menambahkan isi list tetapi dengan urutan yang berbeda ntah di tengah atau di depan bisa gunakan insert
```

```python
bulan.remove("maret")
print(bulan)
#%%%%%%%%%%%%%%%%%%%%%%%%%%%
#untuk menghilangkan isi bisa gunakan remove
```

```python
print("maret" in bulan)
#%%%%%%%%%%%%%%%%%%%%%%%%%%%
#karena maret sudah dihapus maka sudah tidak ada lagi di list yang baru
```

```python
print("januari" in bulan)
```

```python
print(len(bulan))
#%%%%%%%%%%%%%%%%%%%%%%%%%%%
#len digunakan untuk menghitung jumlah kata atau list yang ada
```

```python
#FOR LOOP

zodiak = ["taurus", "libra", "leo", "aquarius"]
for jenis in zodiak:
    print(jenis)
#%%%%%%%%%%%%%%%%%%%%%%%%%%
#for loop bisa digunakan untuk menampilkan semua isi yang ada pada list
#untuk menghasilkan semua isi list tidak hanya bisa dengan menggunakan for loop saja tetapi bisa juga menggunakan while 
#untuk penggunaan while atau for loop itu nanti tergantung jenis data yang akan diolah
```

```python
#FUNGSI RANGE

banyak = range(10)
print(banyak)
#%%%%%%%%%%%%%%%%%%%%%%%%%%
# range digunakan untuk bilangan berurutan
```

```python
for isi in banyak:
    print(isi)
#%%%%%%%%%%%%%%%%%%%%%%%%%
#bila digunakan untuk bilangan yang berurutan seperti ini
```

```python
for isi in range(3,8):
    print(isi)
```

```python
for isi in range(0,10,3):
    print(isi)
#%%%%%%%%%%%%%%%%%%%%%%%%%%%
#ini adalah bila range digunakan untuk bilangan berpola
```
