```python
#TIPE DATA PADA PHYTON
#%%%%%%%%%%%%%%%%%%%%%%%%%%%
umur_anda = input("berapakah umur anda sekarang ? ")
umur_ibu = input ("berapakah umur ibu anda sekarang? ")
selisih = int(umur_ibu) - int(umur_anda)
print ("selisih umur anda dan ibu anda adalah " + str (selisih))
```

```python
umur_anda = int(input("berapakah umur anda sekarang ? "))
umur_ibu = int(input ("berapakah umur ibu anda sekarang? "))
print ("selisih umur anda dan ibu anda adalah " + str (umur_ibu - umur_anda))
#%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
#int bisa digunakan di awal seperti di atas
```
